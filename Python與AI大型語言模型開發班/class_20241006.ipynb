```python
import tkinter as tk

def calculate():
    try:
        num1 = float(ent_num1.get())
        num2 = float(ent_num2.get())
        result = num1 + num2
        label_result.config(text=f"結果: {result}")
    except ValueError:
        label_result.config(text="請輸入有效的數字")

root = tk.Tk()
root.title("簡單加法計算器")

# 輸入框和標籤
tk.Label(root, text="數字 1:").grid(row=0, column=0, sticky=tk.W, padx=5, pady=5)
ent_num1 = tk.Entry(root, width=20)
ent_num1.grid(row=0, column=1, padx=5, pady=5)

tk.Label(root, text="數字 2:").grid(row=1, column=0, sticky=tk.W, padx=5, pady=5)
ent_num2 = tk.Entry(root, width=20)
ent_num2.grid(row=1, column=1, padx=5, pady=5)

# 等號按鈕
tk.Button(root, text="=", command=calculate, width=15).grid(row=2, column=0, columnspan=2, pady=10)

# 結果顯示
label_result = tk.Label(root, text="結果: ")
label_result.grid(row=3, column=0, columnspan=2, pady=5)

root.mainloop()
```

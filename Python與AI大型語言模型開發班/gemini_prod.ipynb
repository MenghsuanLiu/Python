```python
import tkinter as tk
from tkinter import scrolledtext
import google.generativeai as genai
from google.cloud import texttospeech
import pygame
import os
import io


os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "./config/service-key.json"
# # 設置 Gemini API 金鑰


# 初始化 Gemini 模型
model = genai.GenerativeModel(model_name = "gemini-1.5-flash-8b-exp-0924")

# 初始化 Text-to-Speech 客戶端
tts_client = texttospeech.TextToSpeechClient()

# 初始化 pygame 用於音頻播放
pygame.init()
pygame.mixer.init()

def send_text():
    user_input = input_text.get("1.0", tk.END).strip()
    if user_input:
        response = model.generate_content(user_input)
        display_result(response.text)

def speak_result():
    text = result_text.get("1.0", tk.END).strip()
    if text:
        # 設置語音合成請求
        synthesis_input = texttospeech.SynthesisInput(text = text)
        
        # 設置語音參數
        voice = texttospeech.VoiceSelectionParams(
            language_code="zh-TW"
            # name="zh-TW-Standard-A"
        )
        
        # 設置音頻配置
        audio_config = texttospeech.AudioConfig(
            audio_encoding = texttospeech.AudioEncoding.MP3
        )
        
        # 執行語音合成
        response = tts_client.synthesize_speech(
            input = synthesis_input, voice = voice, audio_config = audio_config
        )
        
        # 播放合成的音頻
        with io.BytesIO(response.audio_content) as audio_file:
            pygame.mixer.music.load(audio_file)
            pygame.mixer.music.play()
# def send_voice():
#     recognizer = sr.Recognizer()
#     with sr.Microphone() as source:
#         audio = recognizer.listen(source)
#     try:
#         text = recognizer.recognize_google(audio, language="zh-TW")
#         input_text.delete("1.0", tk.END)
#         input_text.insert(tk.END, text)
#         response = model.generate_content(text)
#         display_result(response.text)
#     except sr.UnknownValueError:
#         display_result("無法識別語音")
#     except sr.RequestError:
#         display_result("無法連接到語音識別服務") 

def display_result(text):
    result_text.config(state=tk.NORMAL)
    result_text.delete("1.0", tk.END)
    result_text.insert(tk.END, text)
    result_text.config(state=tk.DISABLED)

# 創建主窗口
root = tk.Tk()
root.title("Gemini AI 助手")

# 創建輸入框
input_text = scrolledtext.ScrolledText(root, height=5)
input_text.pack(padx=10, pady=10)

# 創建按鈕框架
button_frame = tk.Frame(root)
button_frame.pack(pady=5)

# 創建發送文字按鈕
send_text_button = tk.Button(button_frame, text="發送文字", command=send_text)
send_text_button.pack(side=tk.LEFT, padx=5)

# # 創建語音輸入按鈕
# send_voice_button = tk.Button(button_frame, text="語音輸入", command=send_voice)
# send_voice_button.pack(side=tk.LEFT, padx=5)

# 創建結果顯示區域
result_text = scrolledtext.ScrolledText(root, height=15, state=tk.DISABLED)
result_text.pack(padx=10, pady=10)

# 創建朗讀結果按鈕
speak_button = tk.Button(root, text="朗讀結果", command=speak_result)
speak_button.pack(pady=5)

root.mainloop()
```

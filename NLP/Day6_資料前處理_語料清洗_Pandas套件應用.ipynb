# 資料合併

```python
df1 = pd.DataFrame(np.array([[1, 2, 3], [1, 2, 3]]), columns=["Col1", "Col2", "Col3"])
print(display(df1))
df2 = pd.DataFrame(np.array([[6, 7, 8], [6, 8, 8]]), columns=["Col1", "Col2", "Col3"])
print(display(df2))
pd.concat([df1, df2], axis=0)
```

```python
df1 = pd.DataFrame(np.array([[1, 2, 3], [1, 2, 3]]), columns=["Col1", "Col2", "Col3"])
print(display(df1))
df2 = pd.DataFrame(np.array([[6, 7, 8], [6, 8, 8]]), columns=["Col4", "Col5", "Col6"])
print(display(df2))
pd.concat([df1, df2], axis=1)
```

```python
df1 = pd.DataFrame(np.array([["Andy", 2, 1, 3], ["Tom", 3, 2, 1]]), columns=["Primary_Key", "Col1", "Col2", "Col3"])
print(display(df1))
df2 = pd.DataFrame(np.array([["Andy", 3, 65, 1], ["Tom", 3, 69, 17]]), columns=["Primary_Key", "Col4", "Col5", "Col6"])
print(display(df2))
pd.merge(left=df1, right=df2, on="Primary_Key", how="inner")
```

# 資料轉換

```python
df = pd.read_excel("data_language.xlsx")
print(display(df.head()["text"]))
df["text_replace"] = df["text"].apply(lambda x: x.replace("：", "-----"))
df["text_replace"].head()
```

```python
df = pd.read_excel("data_language.xlsx")
print(display(df.head()["text"]))
df["text_replace"] = df["text"].str.replace("：", "-----")
df["text_replace"].head()
```

### 查看特別函數的微分的圖形，可以一起複習極大極小值，我們會發現極值真的出現在微分等於 0 或是邊界

1. $$f(x)=e^x$$
2. $$g(x)=1/x$$
3. $$h(x)=x^4$$
4. $$\tau(x)=lnx$$

```python
def f(x):
    return np.exp(x)

def g(x):
    return 1/x

def h(x):
    return x**4

def τ(x):
    return np.log(x)
```

```python
# 微分的定義(詳細理由在最後一章會告訴大家)

def diff(fcn):
    h=1e-5
    return lambda x:(fcn(x+h)-fcn(x))/h
```

#### 為 diff 稍作解釋:

上述 diff 函式的輸入為想微分的函數，例如輸入為前面定義的
$$f(x)=e^x$$
輸出為微分後的函數(還未代入數字)。例如 :

```python
f_prime = diff(f)
```

上一欄為
$$f'(x)=e^x$$
我們可以代入數字檢驗，又因為此函數微分前後的值不變所以檢測項目為 :
把任意數字代入 f(x) 及 f_prime(x)，要產生出一樣(相近，因為電腦可能有誤差，最後一章會解說)的數字 :

```python
f(2) - f_prime(2)
```

#### 實際查看各個函數與微分後之圖形

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5)

fig, ax = plt.subplots(2,2)

ax[0,0].plot(x,f(x))
ax[0,1].plot(x,g(x))
ax[1,0].plot(x,h(x))
ax[1,1].plot(x,τ(x))

ax[0,0].set_title("e^x")
ax[0,1].set_title("1/x")
ax[1,0].set_title("x^4")
ax[1,1].set_title("lnx")  # 注意 log 在小於 0 的地方沒有定義

fig.tight_layout()
plt.show()
```

```python
f_prime = diff(f)
g_prime = diff(g)
h_prime = diff(h)
τ_prime = diff(τ)
```

```python
x=np.linspace(-5,5)

fig, ax = plt.subplots(2,2)

ax[0,0].plot(x,f_prime(x))
ax[0,1].plot(x,g_prime(x))
ax[1,0].plot(x,h_prime(x))
ax[1,1].plot(x,τ_prime(x))

ax[0,0].set_title("d(e^x)/dx")
ax[0,1].set_title("d(1/x)/dx")
ax[1,0].set_title("d(x^4)/dx")
ax[1,1].set_title("(lnx)/dx")  # 注意 log 在小於 0 的地方沒有定義

fig.tight_layout()
plt.show()
```
